### Creating DF from the scraped list of yelp and google restaurant lists

#### Basic data exploration

In this step, we are performing below operations to check what the data set comprises of. We are checking the below things:
- head of the dataset
- the shape of the dataset
- info of the dataset
- summary of the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def createYelpDF():
    df = pd.read_csv("./../dataset/yelp/restaurant_list.csv")
    df = df.dropna()
    return df
```

#### Head: The head function will tell you the top records in the data set

```python
df = createYelpDF()
df.head()
```

#### Tail: The tail function will tell you the bottom records in the dataset

```python
df.tail()
```

## Checking for Duplicate Records

```python
duplicate = df.duplicated()
print(duplicate.sum())
df[duplicate]
```

## Bar Plot for review counts

```python
df.plot(kind='bar', x="name", y="review_count", figsize=(90, 80))
plt.title('Number of reviews by restaurants')
plt.ylabel('Review count')
plt.xlabel('Name')
```

## Bar Plot for Ratings

```python
df.plot(kind='bar', x="name", y="rating", figsize=(90, 10))
plt.title('Rating by restaurants')
plt.ylabel('Rating')
plt.xlabel('Name')
```
